```python
import os
import pandas as pd

from PIL import Image
```

```python
df = pd.read_csv('table_10.csv', index_col=0)
df
```

```python
df00 = pd.DataFrame()

for page, path in df['path00'].items():
    image = Image.open(path)    
    df00.loc[page, 'format'] = image.format
    df00.loc[page, 'width'] = image.width
    df00.loc[page, 'height'] = image.height

df00[['width', 'height']] = df00[['width', 'height']].astype(int)
df00.value_counts()
```
