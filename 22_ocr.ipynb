```python
import subprocess
import pandas as pd
 
from PIL import Image
from pytesseract import image_to_string
```

```python
df10 = pd.read_csv('table_10.csv', index_col=0)

paths = df10[df10['lect'] == 4]['path00'].to_dict()
```

```python
images = {page: Image.open(path)
    for page, path in paths.items()}

black_list = '$’'

config = '--oem 1 --psm 3 -l deu+ell+eng'

string = f'{"."*40}\n'.join([
    image_to_string(image, config=config)
    for page, image in images.items()])
```

```python
subprocess.run('pbcopy', text=True, input=string)
print(string)
```
