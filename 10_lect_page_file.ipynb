```python
import os
import pandas as pd

from collections import OrderedDict
from IPython.display import HTML
```

```python
# Lecture number to first page of the lecture.

lect2first = OrderedDict()

lect2first[ 1] =   1
lect2first[ 2] =  13
lect2first[ 3] =  24
lect2first[ 4] =  33
lect2first[ 5] =  40

lect2first[ 6] =  53
lect2first[ 7] =  62
lect2first[ 8] =  77
lect2first[ 9] =  86
lect2first[10] =  95

lect2first[11] = 109
lect2first[12] = 125
lect2first[13] = 135
lect2first[14] = 150
lect2first[15] = 161

lect2first[16] = 173
lect2first[17] = 197
lect2first[18] = 214
lect2first[19] = 233
lect2first[20] = 251

lect2first[21] = 273
lect2first[22] = 290
lect2first[23] = 308
lect2first[24] = 322
lect2first[25] = 347

lect2first[26] = 369
lect2first[27] = 388
lect2first[28] = 406
lect2first[29] = 428
lect2first[30] = 449

last = 465
```

```python
# Lecture number to list of all pages in the lecture.

lect2pages = OrderedDict()

for lect, first in lect2first.items():
    if lect != 30:
        pages = range(first, lect2first[lect + 1])
    else:
        pages = range(first, last + 1)
    lect2pages[lect] = list(pages)
```

```python
# This book has been scanned twice by Google.  Maybe.
# Hence, two tif directories.

dir00 = 'vorlesungenberm00kircgoog_tif'
dir01 = 'vorlesungenberm01kircgoog_tif'
```

```python
# Slice to get just the lectures.
# Remove duplicate / error scans.

## If there is a bad scan and they realize it,
## they rescan but the bad scan remains?  

files00 = sorted(os.listdir(dir00))
files01 = sorted(os.listdir(dir01))

files00 = files00[16:489]
files01 = files01[16:487]

files00.remove('vorlesungenberm00kircgoog_0284.tif')
files00.remove('vorlesungenberm00kircgoog_0285.tif')
files00.remove('vorlesungenberm00kircgoog_0286.tif')
files00.remove('vorlesungenberm00kircgoog_0287.tif')
files00.remove('vorlesungenberm00kircgoog_0288.tif')
files00.remove('vorlesungenberm00kircgoog_0289.tif')
files00.remove('vorlesungenberm00kircgoog_0460.tif')
files00.remove('vorlesungenberm00kircgoog_0461.tif')

files01.remove('vorlesungenberm01kircgoog_0212.tif')
files01.remove('vorlesungenberm01kircgoog_0213.tif')
files01.remove('vorlesungenberm01kircgoog_0254.tif')
files01.remove('vorlesungenberm01kircgoog_0255.tif')
files01.remove('vorlesungenberm01kircgoog_0458.tif')
files01.remove('vorlesungenberm01kircgoog_0459.tif')

assert len(files00) == last
assert len(files01) == last
```

```python
# Page number to file for the scan / tif.

page2file00 = OrderedDict((i + 1, file00) for i, file00 in enumerate(files00))
page2file01 = OrderedDict((i + 1, file01) for i, file01 in enumerate(files01))
```

```python
# Check that the page numbers seem correct.
# We know the starting pages for lectures.
# Display these pages.

start_pages = [pages[0] for lecture, pages in lect2pages.items()]

paths00 = [os.path.join(dir00, page2file00[page]) for page in start_pages]
paths01 = [os.path.join(dir01, page2file01[page]) for page in start_pages]

imgs00 = ''.join([f'<div><img src="{path}"></div>' for path in paths00])
imgs01 = ''.join([f'<div><img src="{path}"></div>' for path in paths01])

style5 = '"display: grid; grid-template-columns: repeat(5, 1fr);"'

html00 = f'<div style={style5}>{imgs00}</div>'
html01 = f'<div style={style5}>{imgs01}</div>'
```

```python
dict00 = OrderedDict({'lect': [], 'page': [], 'file': []})

for lect, pages in lect2pages.items():
    for page in pages:
        dict00['lect'].append(lect)
        dict00['page'].append(page)
        dict00['file'].append(page2file00[page])
        
vorlesungen00 = pd.DataFrame(dict00)
vorlesungen00
```

```python
dict01 = OrderedDict({'lect': [], 'page': [], 'file': []})

for lect, pages in lect2pages.items():
    for page in pages:
        dict01['lect'].append(lect)
        dict01['page'].append(page)
        dict01['file'].append(page2file01[page])
        
vorlesungen01 = pd.DataFrame(dict01)
vorlesungen01
```

```python
vorlesungen00.to_csv('vorlesungen00_lect_page_file.csv', index=False)
vorlesungen01.to_csv('vorlesungen01_lect_page_file.csv', index=False)
```
