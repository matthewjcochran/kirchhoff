```python
import os
import pandas as pd
import pytesseract

from PIL import Image
```

```python
df = pd.read_csv('table_10.csv', index_col=0)
```

```python
lect = 1

paths00 = df.loc[df['lect'] == lect, 'path00']
paths01 = df.loc[df['lect'] == lect, 'path01']

images00 = [Image.open(path) for path in paths00]
images01 = [Image.open(path) for path in paths01]
```

```python
list00 = [pytesseract.image_to_string(image, lang='deu') for image in images00]
list01 = [pytesseract.image_to_string(image, lang='deu') for image in images01]

text00 = '##########\n'.join(list00)
text01 = '##########\n'.join(list01)
```
